```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('health.activity.day_summary.csv')
```

```python
print(df.head())
print(df.describe())
```

```python
plt.figure(figsize=(10, 6))
```

```python
sns.scatterplot(x='Steps', y='Calories', data=df)
```

```python
plt.title('Relationship Between Step Count and Calorie Consumption')
```

```python
plt.xlabel('Step Count')
```

```python
plt.ylabel('Calorie Consumption')
```

```python
plt.show()
```

```python
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(12, 6))
```

```python
sns.lineplot(x='update_time', y='exercise_time', data=df, marker='o')
```

```python
plt.title('Exercise Time Over Time')
```

```python
plt.xlabel('Date')
```

```python
plt.ylabel('Exercise Time (minutes)')
```

```python
plt.xticks(rotation=45)
```
